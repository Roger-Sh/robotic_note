```python
# 导入需要使用的库,其中numpy是支持数组和矩阵运算的科学计算库,而matplotlib是绘图库
import numpy as np
import matplotlib.pyplot as plt


class BernoulliBandit:
    """ 
    伯努利多臂老虎机,输入K表示拉杆个数 
    伯努利分布：每次拉下拉杆有 p 的概率获得的奖励为 1，有(1-p)的概率获得的奖励为 0
    奖励为 1 代表获奖，奖励为 0 代表没有获奖
    """
    
    def __init__(self, K):
        # K表示拉杆个数 
        self.K = K 
        
        # 随机生成K个0～1的数,作为拉动每根拉杆的获奖概率
        self.probs = np.random.uniform(size=K)  
        
        # 概率
        self.best_idx = np.argmax(self.probs)  # 获奖概率最大的拉杆
        self.best_prob = self.probs[self.best_idx]  # 最大的获奖概率
        

    def step(self, k):
        # 当玩家选择了k号拉杆后,根据拉动该老虎机的k号拉杆获得奖励的概率返回1（获奖）或0（未
        # 获奖）
        if np.random.rand() < self.probs[k]:
            return 1
        else:
            return 0

# 设定随机种子,使实验具有可重复性
np.random.seed(1)  

# 设置10个拉杆的老虎机
K = 10
bandit_10_arm = BernoulliBandit(K)
print("随机生成了一个%d臂伯努利老虎机" % K)
print("获奖概率最大的拉杆为%d号,其获奖概率为%.4f" %
      (bandit_10_arm.best_idx, bandit_10_arm.best_prob))

# 随机生成了一个10臂伯努利老虎机
# 获奖概率最大的拉杆为1号,其获奖概率为0.7203
```

```python
class Solver:
    """ 
    多臂老虎机算法基本框架 
    根据策略选择动作、
    根据动作获取奖励、
    更新期望奖励估值、
    更新累积懊悔和计数
    """
    
    def __init__(self, bandit):
        """
        初始化 solver
        """
        
        # 获取多臂老虎机
        self.bandit = bandit
        
        # 每根拉杆的尝试次数
        self.counts = np.zeros(self.bandit.K)  
        
        # 当前步的累积懊悔
        self.regret = 0.  
        
        # 维护一个列表,记录每一步的动作
        self.actions = []  
        
        # 维护一个列表,记录每一步的累积懊悔
        self.regrets = []  

    def update_regret(self, k):
        # 计算累积懊悔并保存,k为本次动作选择的拉杆的编号
        # 这里的懊悔指的是当前的拉的杆子的获奖概率与最大获奖概率的杆子之间的差距
        self.regret += self.bandit.best_prob - self.bandit.probs[k]
        
        # 保存到懊悔列表中
        self.regrets.append(self.regret)

    def run_one_step(self):
        # 返回当前动作选择哪一根拉杆,由每个具体的策略实现
        # 需要继承实现
        raise NotImplementedError

    def run(self, num_steps):
        # 运行一定次数,num_steps为总运行次数
        for _ in range(num_steps):
            # 选择一根杆子作为动作
            k = self.run_one_step()
            
            # 该动作的计数器+1
            self.counts[k] += 1
            
            # 保存该动作到动作列表
            self.actions.append(k)
            
            # 更新懊悔列表
            self.update_regret(k)
```

```python
class EpsilonGreedy(Solver):
    """ 
    epsilon贪婪算法,继承Solver类
    需要实现 run_one_step
    
    ϵ-贪心算法
    概率 1-ϵ 选择以往经验中期望奖励估值最大的拉杆
    概率 ϵ 选择随便一根拉杆 
    """
    
    def __init__(self, bandit, epsilon=0.01, init_prob=1.0):
        # 调用Solver的__init__
        super(EpsilonGreedy, self).__init__(bandit)
        
        # 概率 ϵ
        self.epsilon = epsilon
        
        #初始化拉动所有拉杆的期望奖励估值，默认为 1.0
        self.estimates = np.array([init_prob] * self.bandit.K)

    def run_one_step(self):
        
        # 概率 ϵ 选择随便一根拉杆
        if np.random.random() < self.epsilon:
            k = np.random.randint(0, self.bandit.K)  
            
        # 概率 1-ϵ 选择以往经验中期望奖励估值最大的拉杆
        else:
            k = np.argmax(self.estimates)
            
        # 得到本次动作的奖励
        r = self.bandit.step(k)
        
        # 更新该动作的期望奖励估值
        self.estimates[k] += 1. / (self.counts[k] + 1) * (r - self.estimates[k])
        return k
```

```python
def plot_results(solvers, solver_names):
    """
    生成累积懊悔随时间变化的图像。
    输入solvers是一个列表,列表中的每个元素是一种特定的策略。
    而solver_names也是一个列表,存储每个策略的名称
    """
    for idx, solver in enumerate(solvers):
        time_list = range(len(solver.regrets))
        plt.plot(time_list, solver.regrets, label=solver_names[idx])
    plt.xlabel('Time steps')
    plt.ylabel('Cumulative regrets')
    plt.title('%d-armed bandit' % solvers[0].bandit.K)
    plt.legend()
    plt.show()

# 设定随机种子,使实验具有可重复性
np.random.seed(1)

# 将10臂老虎机和epsilon输入EpsilonGreedy求解器
epsilon_greedy_solver = EpsilonGreedy(bandit_10_arm, epsilon=0.01)

# 运行5000步
epsilon_greedy_solver.run(5000)
print('epsilon-贪婪算法的累积懊悔为：', epsilon_greedy_solver.regret)
plot_results([epsilon_greedy_solver], ["EpsilonGreedy"])

# epsilon-贪婪算法的累积懊悔为：25.526630933945313
```

```python
# 设置不同的随机种子
np.random.seed(0)

# 设置不同的epsilon贪婪求解器
# 更大的epsilon说明每次更可能随机选择一个杆子
epsilons = [1e-4, 0.01, 0.1, 0.25, 0.5]
epsilon_greedy_solver_list = [
    EpsilonGreedy(bandit_10_arm, epsilon=e) for e in epsilons
]
epsilon_greedy_solver_names = ["epsilon={}".format(e) for e in epsilons]

# 每个求解器跑5000次
for solver in epsilon_greedy_solver_list:
    solver.run(5000)

# 打印累计懊悔
plot_results(epsilon_greedy_solver_list, epsilon_greedy_solver_names)

# 基本上无论 epsilon 取值多少，累积懊悔都是线性增长的
```

```python
class DecayingEpsilonGreedy(Solver):
    """ 
    epsilon值随时间衰减的epsilon-贪婪算法,继承Solver类
    """
    
    def __init__(self, bandit, init_prob=1.0):
        # 求解器初始化
        super(DecayingEpsilonGreedy, self).__init__(bandit)
        
        # 初始化拉动所有拉杆的期望奖励估值，默认为 1.0
        self.estimates = np.array([init_prob] * self.bandit.K)
        
        # 记录整体的动作计数器
        self.total_count = 0

    def run_one_step(self):
        """
        执行一次动作
        """
        
        # 计数器+1
        self.total_count += 1
        
        # epsilon值随时间衰减
        # 这里 epsilon值为总体动作计数器的倒数，动作执行越多，epsilon值越小，随机动作的概率越低
        if np.random.random() < 1 / self.total_count:  
            k = np.random.randint(0, self.bandit.K)
            
        # 1-epsilon 的概率，抽取目前最大回报的杆子
        else:
            k = np.argmax(self.estimates)

        # 执行一次动作，获得回报
        r = self.bandit.step(k)
        
        # 更新该动作的期望奖励估值
        self.estimates[k] += 1. / (self.counts[k] + 1) * (r - self.estimates[k])
        return k


# 设置随机种子
np.random.seed(1)

# 初始化 DecayingEpsilonGreedy
decaying_epsilon_greedy_solver = DecayingEpsilonGreedy(bandit_10_arm)
decaying_epsilon_greedy_solver.run(5000)
print('epsilon值衰减的贪婪算法的累积懊悔为：', decaying_epsilon_greedy_solver.regret)

# 累计懊悔会随着时间增加趋于稳定，说明随着时间增加，策略中的随机性减少，更能找到最佳动作
# 随时间做反比例衰减的 epsilon-贪婪算法能够使累积懊悔与时间步的关系变成次线性
plot_results([decaying_epsilon_greedy_solver], ["DecayingEpsilonGreedy"])

# epsilon值衰减的贪婪算法的累积懊悔为：10.114334931260183
```

```python
class UCB(Solver):
    """ 
    UCB算法,继承Solver类 
    上置信界（upper confidence bound，UCB）
    引入不确定性度量, 随着一个动作被尝试次数的增加而减小
    用来解决某个杆子被执行的次数过少，导致其回报的不确定性太大的问题
    参考：霍夫丁不等式
    
    UCB 算法在每次选择拉杆前，先估计每根拉杆的期望奖励的上界UCB，
    使得拉动每根拉杆的期望奖励只有一个较小的概率超过这个上界，
    接着选出期望奖励上界最大的拉杆，从而选择最有可能获得最大期望奖励的拉杆
    
    """
    def __init__(self, bandit, coef, init_prob=1.0):
        # 初始化solver
        super(UCB, self).__init__(bandit)
        
        # 动作计数器
        self.total_count = 0
        
        # 初始化各个动作的期望
        self.estimates = np.array([init_prob] * self.bandit.K)
        
        # UCB的更新系数，不确定性比重的系数
        self.coef = coef

    def run_one_step(self):
        """
        执行动作
        """
        
        # 动作计数器+1
        self.total_count += 1
        
        # 计算上置信界， 即不确定性度量
        # 某个拉杆的执行次数越多，该拉杆的不确定性度量就越小
        # 该上置信界同时还考虑了每个动作的期望奖励估值
        # 所以最终选择动作时，即考虑到这个动作的不确定性的大小，越大越容易被选中
        # 也考虑每个动作的期望奖励估值，期望奖励越大，越容易被选中
        # 一开始不确定性度量起到的作用更大
        # 等所有的动作的不确定性度量都小了之后，则每个动作本身的期望奖励估值起到的作用更大
        ucb = self.estimates + self.coef * np.sqrt(
            np.log(self.total_count) / (2 * (self.counts + 1)))  
        
        # 选出上置信界最大的拉杆，即最不确定的拉杆
        k = np.argmax(ucb)
        
        # 执行该动作
        r = self.bandit.step(k)
        
        # 更新该动作的期望奖励估值
        self.estimates[k] += 1. / (self.counts[k] + 1) * (r - self.estimates[k])
        return k


# 随机种子
np.random.seed(1)
# 控制不确定性比重的系数
coef = 1  

# 初始化UCB求解器
UCB_solver = UCB(bandit_10_arm, coef)

# 运行5000次
UCB_solver.run(5000)

# 上置信界算法的累积懊悔为： 70.45281214197854
print('上置信界算法的累积懊悔为：', UCB_solver.regret)

# 打印累计懊悔
# 累计懊悔同样会收敛
# 收敛速度会比衰减epsilon贪婪算法慢，因为会考虑各个动作的上置信界，
# 使每个动作的不确定性都降低后，才会更准确选择期望奖励估值更大的动作
plot_results([UCB_solver], ["UCB"])

```

```python
class ThompsonSampling(Solver):
    """ 
    汤普森采样算法,继承Solver类
    
    汤普森采样算法使用采样的方式，即根据当前每个动作 a 的奖励概率分布进行一轮采样，
    得到一组各根拉杆的奖励样本，再选择样本中奖励最大的动作。
    可以看出，汤普森采样是一种计算所有拉杆的最高奖励概率的蒙特卡洛采样方法。
    """
    def __init__(self, bandit):
        # 初始化老虎机
        super(ThompsonSampling, self).__init__(bandit)
        
        # 列表,表示每根拉杆奖励为1的次数
        self._a = np.ones(self.bandit.K)  
        
        # 列表,表示每根拉杆奖励为0的次数
        self._b = np.ones(self.bandit.K)  

    def run_one_step(self):
        """
        执行一次动作
        """
        
        # 按照Beta分布采样一组奖励样本
        samples = np.random.beta(self._a, self._b)
        
        # 选出采样奖励最大的拉杆
        k = np.argmax(samples)
        
        # 根据动作计算回报
        r = self.bandit.step(k)

        # 更新Beta分布的参数
        self._a[k] += r        # 更新Beta分布的第一个参数 表示每根拉杆奖励为1的次数
        self._b[k] += (1 - r)  # 更新Beta分布的第二个参数 表示每根拉杆奖励为0的次数
        return k


np.random.seed(1)
thompson_sampling_solver = ThompsonSampling(bandit_10_arm)
thompson_sampling_solver.run(5000)
print('汤普森采样算法的累积懊悔为：', thompson_sampling_solver.regret)

# 打印累计懊悔
# 累计懊悔同样会收敛
plot_results([thompson_sampling_solver], ["ThompsonSampling"])

# 汤普森采样算法的累积懊悔为：57.19161964443925
```
