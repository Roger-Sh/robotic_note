# 第8章 DQN改进算法

```python
import random
import gym
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import rl_utils
from tqdm import tqdm


class Qnet(torch.nn.Module):
    ''' 
    只有一层隐藏层的Q网络 
    '''
    
    def __init__(self, state_dim, hidden_dim, action_dim):
        """
        初始化Q网络
        state_dim, hidden_dim, action_dim
        """
        
        # 调用 torch.nn.Module 的初始化函数
        super(Qnet, self).__init__()
        
        # 创建Q网络
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, action_dim)

    def forward(self, x):
        """
        Q网络前向传播
        """
        
        x = F.relu(self.fc1(x))
        return self.fc2(x)
```

```python
class DQN:
    ''' 
    DQN算法,包括Double DQN 
    '''
    
    def __init__(self,
                 state_dim,
                 hidden_dim,
                 action_dim,
                 learning_rate,
                 gamma,
                 epsilon,
                 target_update,
                 device,
                 dqn_type='VanillaDQN'):
        """
        初始化DQN
        
        state_dim, 状态维度
        hidden_dim, 隐藏层维度
        action_dim, 动作维度
        learning_rate, 学习率
        gamma, 折扣因子
        epsilon, e-greedy 概率参数
        target_update, 目标网络更新间隔轮数
        device, 训练硬件
        dqn_type='VanillaDQN', DQN类型, "VanillaDQN", "DoubleDQN"
        
        """
        
        # 创建训练网络 q_net
        self.action_dim = action_dim
        self.q_net = Qnet(state_dim, hidden_dim, self.action_dim).to(device)
        
        # 创建目标网络 target_q_net
        self.target_q_net = Qnet(state_dim, hidden_dim, self.action_dim).to(device)
        
        # ADAM优化器
        self.optimizer = torch.optim.Adam(self.q_net.parameters(),
                                          lr=learning_rate)
        
        # 初始化参数
        self.gamma = gamma
        self.epsilon = epsilon
        self.target_update = target_update
        self.dqn_type = dqn_type
        self.device = device
        
        # 训练计数器
        self.count = 0

    def take_action(self, state):
        """
        e-greedy 动作策略
        
        基于q_net的推理结果
        """
        
        # <e，随机选择动作
        if np.random.random() < self.epsilon:
            action = np.random.randint(self.action_dim)
        
        # >e，通过q_net选择最大价值的动作
        else:
            state = torch.tensor([state], dtype=torch.float).to(self.device)
            action = self.q_net(state).argmax().item()
            
        return action

    def max_q_value(self, state):
        """
        获取q_net的动作最大价值
        """
        
        state = torch.tensor([state], dtype=torch.float).to(self.device)
        return self.q_net(state).max().item()

    def update(self, transition_dict):
        """
        训练更新网络参数
        """
        
        # 获取当前批次训练数据的各个元素
        states = torch.tensor(transition_dict['states'],
                              dtype=torch.float).to(self.device)
        actions = torch.tensor(transition_dict['actions']).view(-1, 1).to(
            self.device)
        rewards = torch.tensor(transition_dict['rewards'],
                               dtype=torch.float).view(-1, 1).to(self.device)
        next_states = torch.tensor(transition_dict['next_states'],
                                   dtype=torch.float).to(self.device)
        dones = torch.tensor(transition_dict['dones'],
                             dtype=torch.float).view(-1, 1).to(self.device)

        # 基于当前状态，通过q_net计算对应动作的Q值
        q_values = self.q_net(states).gather(1, actions)  
        
        # 下个状态的最大Q值
        if self.dqn_type == 'DoubleDQN': 
            # DoubleDQN
            # 通过 q_net 选择出最大价值的动作
            max_action = self.q_net(next_states).max(1)[1].view(-1, 1) 
            # 通过 target_q_net 计算选择的动作的下一个状态中的价值
            max_next_q_values = self.target_q_net(next_states).gather(1, max_action)
        else: 
            # 原始DQN，Vanilla DQN
            # 通过target_q_net直接选出下一状态的最大价值动作
            max_next_q_values = self.target_q_net(next_states).max(1)[0].view(-1, 1)
            
        # TD误差目标
        q_targets = rewards + self.gamma * max_next_q_values * (1 - dones)  
        
        # 均方误差损失函数
        dqn_loss = torch.mean(F.mse_loss(q_values, q_targets))  
        
        # PyTorch中默认梯度会累积,这里需要显式将梯度置为0
        self.optimizer.zero_grad()  
        
        # 反向传播更新参数
        dqn_loss.backward()  
        
        # 优化器迭代
        self.optimizer.step()

        # 更新目标网络target_q_net的参数
        if self.count % self.target_update == 0:
            self.target_q_net.load_state_dict(self.q_net.state_dict()) 
            
        # 更新训练计数器
        self.count += 1
```

```python
# 配置训练参数
lr = 1e-2
num_episodes = 200
hidden_dim = 128
gamma = 0.98
epsilon = 0.01
target_update = 50
buffer_size = 5000
minimal_size = 1000
batch_size = 64
device = torch.device("cuda") if torch.cuda.is_available() else torch.device(
    "cpu")

# 设置训练环境
env_name = 'Pendulum-v1'
env = gym.make(env_name)
state_dim = env.observation_space.shape[0] # 状态维度
action_dim = 11  # 动作维度，将连续动作分成11个离散动作


def dis_to_con(discrete_action, env, action_dim):  
    """
    离散动作转回连续的函数
    """

    action_lowbound = env.action_space.low[0]  # 连续动作的最小值
    action_upbound = env.action_space.high[0]  # 连续动作的最大值
    
    # 通过离散编号计算出动作对应的连续值
    return action_lowbound + (discrete_action / (action_dim - 1)) * (action_upbound - action_lowbound)
```

```python
def train_DQN(agent, env, num_episodes, replay_buffer, minimal_size,
              batch_size):
    """
    训练DQN
    
    agent, 智能体
    env, 环境
    num_episodes, 训练轮数 
    replay_buffer, 经验回放池
    minimal_size, 经验回放池最小尺寸
    batch_size，采样批次大小
    """
    
    # 初始化训练资源
    return_list = [] # 回报历史记录
    max_q_value_list = [] # 最大价值历史记录
    max_q_value = 0 # 当前最大价值
    
    # 可视化进度条
    for i in range(10):
        with tqdm(total=int(num_episodes / 10), desc='Iteration %d' % i) as pbar:
            # 训练循环
            for i_episode in range(int(num_episodes / 10)):
                # 初始化当前训练资源
                episode_return = 0
                state = env.reset()
                done = False
                
                # 开始训练
                while not done:
                    # 选择动作，通过q_net
                    action = agent.take_action(state)
                    
                    # 平滑处理最大q值
                    max_q_value = agent.max_q_value(state) * 0.005 + max_q_value * 0.995  
                    
                    # 保存每个状态的最大Q值
                    max_q_value_list.append(max_q_value)  
                    
                    # 将动作转回连续值
                    action_continuous = dis_to_con(action, env, agent.action_dim)
                    
                    # 执行动作，获得下一状态，回报，结束标识
                    next_state, reward, done, _ = env.step([action_continuous])
                    
                    # 加入经验回放池
                    replay_buffer.add(state, action, reward, next_state, done)
                    
                    # 更新状态
                    state = next_state
                    
                    # 累计奖励
                    episode_return += reward
                    
                    # 经验回放池中采样数据，进行参数更新
                    if replay_buffer.size() > minimal_size:
                        # 采样
                        b_s, b_a, b_r, b_ns, b_d = replay_buffer.sample(
                            batch_size)
                        transition_dict = {
                            'states': b_s,
                            'actions': b_a,
                            'next_states': b_ns,
                            'rewards': b_r,
                            'dones': b_d
                        }
                        
                        # 更新
                        agent.update(transition_dict)
                        
                # 记录回报
                return_list.append(episode_return)
                
                # 打印训练结果
                if (i_episode + 1) % 10 == 0:
                    pbar.set_postfix({
                        'episode':
                        '%d' % (num_episodes / 10 * i + i_episode + 1),
                        'return':
                        '%.3f' % np.mean(return_list[-10:])
                    })
                pbar.update(1)
                
                
    return return_list, max_q_value_list
```

```python
# random
random.seed(0)
np.random.seed(0)
env.seed(0)
torch.manual_seed(0)

# replay_buffer
replay_buffer = rl_utils.ReplayBuffer(buffer_size)

# agent，原始DQN
agent = DQN(state_dim, hidden_dim, action_dim, lr, gamma, epsilon,
            target_update, device)

# train_DQN
return_list, max_q_value_list = train_DQN(agent, env, num_episodes,
                                          replay_buffer, minimal_size,
                                          batch_size)

# 可视化
episodes_list = list(range(len(return_list)))
mv_return = rl_utils.moving_average(return_list, 5)
plt.plot(episodes_list, mv_return)
plt.xlabel('Episodes')
plt.ylabel('Returns')
plt.title('DQN on {}'.format(env_name))
plt.show()

frames_list = list(range(len(max_q_value_list)))
plt.plot(frames_list, max_q_value_list)
plt.axhline(0, c='orange', ls='--')
plt.axhline(10, c='red', ls='--')
plt.xlabel('Frames')
plt.ylabel('Q value')
plt.title('DQN on {}'.format(env_name))
plt.show()

# DQN 算法在倒立摆环境中能取得不错的回报，
# 最后的期望回报在-200 左右，但是不少Q值超过了 0，有一些还超过了 10，
# 该现象便是 DQN 算法中的Q值过高估计
```

```python
# random
random.seed(0)
np.random.seed(0)
env.seed(0)
torch.manual_seed(0)

# replay_buffer
replay_buffer = rl_utils.ReplayBuffer(buffer_size)

# agent, doubleDQN
agent = DQN(state_dim, hidden_dim, action_dim, lr, gamma, epsilon,
            target_update, device, 'DoubleDQN')

# train_DQN
return_list, max_q_value_list = train_DQN(agent, env, num_episodes,
                                          replay_buffer, minimal_size,
                                          batch_size)

# 可视化
episodes_list = list(range(len(return_list)))
mv_return = rl_utils.moving_average(return_list, 5)
plt.plot(episodes_list, mv_return)
plt.xlabel('Episodes')
plt.ylabel('Returns')
plt.title('Double DQN on {}'.format(env_name))
plt.show()

frames_list = list(range(len(max_q_value_list)))
plt.plot(frames_list, max_q_value_list)
plt.axhline(0, c='orange', ls='--')
plt.axhline(10, c='red', ls='--')
plt.xlabel('Frames')
plt.ylabel('Q value')
plt.title('Double DQN on {}'.format(env_name))
plt.show()

# 与普通的 DQN 相比，Double DQN 比较少出现Q值大于 0 的情况，
# 说明Q值过高估计的问题得到了很大缓解。
```

```python
class VAnet(torch.nn.Module):
    ''' 
    DuelingNet
    只有一层隐藏层的A网络和V网络 
    '''
    
    def __init__(self, state_dim, hidden_dim, action_dim):
        """
        初始化DuelingNet的结构
        
        state_dim, hidden_dim, action_dim
        """
        
        # 调用 torch.nn.Module 的初始化函数
        super(VAnet, self).__init__()
        
        # 共享网络部分，隐藏层
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)  
        
        # 输出层：优势函数 A 部分
        self.fc_A = torch.nn.Linear(hidden_dim, action_dim)
        # 输出层：状态价值函数 V 部分
        self.fc_V = torch.nn.Linear(hidden_dim, 1)

    def forward(self, x):
        """
        前向传播
        """
        
        A = self.fc_A(F.relu(self.fc1(x)))
        V = self.fc_V(F.relu(self.fc1(x)))
        
        # Q值由V值和A值计算得到，这里通过获取A的平均值解决V的不唯一性问题
        Q = V + A - A.mean(1).view(-1, 1)
        
        return Q


class DQN:
    ''' 
    DQN算法,包括Double DQN和Dueling DQN 
    '''
    
    def __init__(self,
                 state_dim,
                 hidden_dim,
                 action_dim,
                 learning_rate,
                 gamma,
                 epsilon,
                 target_update,
                 device,
                 dqn_type='VanillaDQN'):
        """
        初始化DQN
        state_dim, 状态维度
        hidden_dim, 隐藏层维度
        action_dim, 动作维度
        learning_rate, 学习率
        gamma, 折扣因子
        epsilon, e-greedy 概率参数
        target_update, 目标网络更新间隔轮数
        device, 训练硬件
        dqn_type='VanillaDQN', DQN类型, "VanillaDQN", "DoubleDQN", "DuelingDQN"
        """
        
        self.action_dim = action_dim
        
        # Dueling DQN采取不一样的网络框架
        if dqn_type == 'DuelingDQN':  
            self.q_net = VAnet(state_dim, hidden_dim,
                               self.action_dim).to(device)
            self.target_q_net = VAnet(state_dim, hidden_dim,
                                      self.action_dim).to(device)
        
        # VanillaDQN 与 DoubleDQN 采用同样的网络结构
        else:
            self.q_net = Qnet(state_dim, hidden_dim,
                              self.action_dim).to(device)
            self.target_q_net = Qnet(state_dim, hidden_dim,
                                     self.action_dim).to(device)
            
        # optimizer
        self.optimizer = torch.optim.Adam(self.q_net.parameters(),
                                          lr=learning_rate)
        
        # 初始化其他参数
        self.gamma = gamma
        self.epsilon = epsilon
        self.target_update = target_update
        self.dqn_type = dqn_type
        self.device = device
        
        # 初始化训练计数器
        self.count = 0

    def take_action(self, state):
        """
        e-greedy 动作策略
        """
        
        if np.random.random() < self.epsilon:
            action = np.random.randint(self.action_dim)
        else:
            state = torch.tensor([state], dtype=torch.float).to(self.device)
            action = self.q_net(state).argmax().item()
            
        return action

    def max_q_value(self, state):
        """
        获取当前状态下通过q_net的最大动作价值
        """
        state = torch.tensor([state], dtype=torch.float).to(self.device)
        return self.q_net(state).max().item()

    def update(self, transition_dict):
        """
        更新网络参数
        transition_dict: 采样的训练数据
        """
        
        # 获取各个训练元素
        states = torch.tensor(transition_dict['states'],
                              dtype=torch.float).to(self.device)
        actions = torch.tensor(transition_dict['actions']).view(-1, 1).to(
            self.device)
        rewards = torch.tensor(transition_dict['rewards'],
                               dtype=torch.float).view(-1, 1).to(self.device)
        next_states = torch.tensor(transition_dict['next_states'],
                                   dtype=torch.float).to(self.device)
        dones = torch.tensor(transition_dict['dones'],
                             dtype=torch.float).view(-1, 1).to(self.device)

        # 通过q_net获取对应动作的Q值
        q_values = self.q_net(states).gather(1, actions)
        
        # DoubleDQN
        # 通过 q_net 选择下一状态时选取的动作
        # 通过 target_q_net 计算下一状态选择的动作对应的价值
        if self.dqn_type == 'DoubleDQN':
            max_action = self.q_net(next_states).max(1)[1].view(-1, 1)
            max_next_q_values = self.target_q_net(next_states).gather(1, max_action)
            
        # VanillaDQN, DuelingDQN
        # 直接通过 target_q_net 选择下一状态的动作和计算动作的价值
        else:
            max_next_q_values = self.target_q_net(next_states).max(1)[0].view(-1, 1)
            
        # TD目标
        q_targets = rewards + self.gamma * max_next_q_values * (1 - dones)
        
        # 通过TD误差计算MSE损失函数
        dqn_loss = torch.mean(F.mse_loss(q_values, q_targets))
        
        # 重置优化器，进行反向传播，迭代优化
        self.optimizer.zero_grad()
        dqn_loss.backward()
        self.optimizer.step()

        # 打印训练过程
        if self.count % self.target_update == 0:
            self.target_q_net.load_state_dict(self.q_net.state_dict())
        
        # 更新训练计数器
        self.count += 1


# random
random.seed(0)
np.random.seed(0)
env.seed(0)
torch.manual_seed(0)

# replay_buffer
replay_buffer = rl_utils.ReplayBuffer(buffer_size)

# agent with DuelingDQN
agent = DQN(state_dim, hidden_dim, action_dim, lr, gamma, epsilon,
            target_update, device, 'DuelingDQN')

# train_DQN
return_list, max_q_value_list = train_DQN(agent, env, num_episodes,
                                          replay_buffer, minimal_size,
                                          batch_size)

# 可视化训练结果
episodes_list = list(range(len(return_list)))
mv_return = rl_utils.moving_average(return_list, 5)
plt.plot(episodes_list, mv_return)
plt.xlabel('Episodes')
plt.ylabel('Returns')
plt.title('Dueling DQN on {}'.format(env_name))
plt.show()

frames_list = list(range(len(max_q_value_list)))
plt.plot(frames_list, max_q_value_list)
plt.axhline(0, c='orange', ls='--')
plt.axhline(10, c='red', ls='--')
plt.xlabel('Frames')
plt.ylabel('Q value')
plt.title('Dueling DQN on {}'.format(env_name))
plt.show()

# 相比于传统的 DQN，Dueling DQN 在多个动作选择下的学习更加稳定，得到的回报最大值也更大
# 随着动作空间的增大，Dueling DQN 相比于 DQN 的优势更为明显
# 可以增加离散动作数（例如 15、25 等），继续进行对比实验。
```
